```python
#Importaciones
import pandas as pd
import matplotlib.pyplot as plt
```

Carga y Lectura de datos

```python
"""Se cargan y se leen los datos de la situación de casos y camas de covid en Bahía Blanca"""
casos_covid_bahia = 'https://raw.githubusercontent.com/manlio99/Materia-de-aprendizaje/master/4_DataWrangling/data/casos_covid_bahia.csv'
camas_covid_bahia = 'https://raw.githubusercontent.com/manlio99/Materia-de-aprendizaje/master/4_DataWrangling/data/camas_covid_bahia.csv'
casos_covid_bahia_df = pd.read_csv(casos_covid_bahia)
camas_covid_bahia_df = pd.read_csv (camas_covid_bahia)
```

# **Situación de casos de covid**

```python
casos_covid_bahia_df.head()
```

```python
"""Se grafican los datos"""
plt.figure()
casos_covid_bahia_df.plot(x="fecha",subplots=True, legend=True, figsize=(15,5), layout=(5,2))
plt.show
```

```python
"""Se realiza una gráfica comparando los casos """
plt.figure()
casos_covid_bahia_df.plot(x="fecha", y=["confirmados","activos", "recuperados","decesos"])
plt.title("Casos", fontdict={"family": "serif", 'weight': 'bold','size': 18})
plt.show()
```

**Observaciones**

* Con el último gráfico se puede tener una mejor representación de la situación, ya que se observa la etapa en la que se encuentran los casos confirmados positivos hasta la fecha.
* También se aprecia como a finales de agosto la curva de confirmados pasa de ser prácticamente lineal a algo más exponencial.

# **Situación de las camas**

```python
camas_covid_bahia_df.head()
```

Se procede a eliminar la columna "informeNumero" y "observaciones", ya que no aportan datos significativos. Y a invertir los datos para que queden acordes a los anteriores.

```python
"""Se quitan las columnas de InformeNumero y Observaciones. Además se invierten los datos"""
if ("informeNumero" or "observaciones") in camas_covid_bahia_df.columns:
  camas_covid_bahia_df=camas_covid_bahia_df.drop(["informeNumero"],axis=1)
  camas_covid_bahia_df=camas_covid_bahia_df.drop(["observaciones"],axis=1)
camas_covid_bahia_df=camas_covid_bahia_df.iloc[::-1]
camas_covid_bahia_df.head()
```

```python
"""Se grafican los datos"""
plt.figure()
camas_covid_bahia_df.plot(x="fecha",subplots=True, legend=True, figsize=(20,10), layout=(6,3))
plt.show()
```

Se grafican algunos datos que resultan de utilidad tenerlos en el mismo gráfico.

```python
"""Se realiza una gráfica de la situación de las camas en los hospitales"""
plt.figure()
camas_covid_bahia_df.plot(x="fecha", y=["total_camas_hospitales","camas_ocupadas_hospitales"])
plt.title("Camas Hospitales", fontdict={"family": "serif", 'weight': 'bold','size': 18})
plt.legend(bbox_to_anchor=(1.05, 1.0), loc="upper left")
plt.show()

"""Se realiza una gráfica de la situación de las camas covid"""
plt.figure()
camas_covid_bahia_df.plot(x="fecha",y=["camas_covid","camas_sospechosos_covid","camas_confirmados_covid"] )
plt.title("Camas covid", fontdict={"family": "serif", 'weight': 'bold','size': 18})
plt.legend(bbox_to_anchor=(1.05, 1.0), loc="upper left")
plt.show()

"""Se realiza una gráfica de la situación de los respiradores"""
plt.figure()
camas_covid_bahia_df.plot(x="fecha",y=["respiradores_totales","respiradores_ocupados"] )
plt.title("Respiradores", fontdict={"family": "serif", 'weight': 'bold','size': 18})
plt.legend(bbox_to_anchor=(1.05, 1.0), loc="upper left")
plt.show()

"""Se realiza una gráfica de la situación de los respiradores covid"""
plt.figure()
camas_covid_bahia_df.plot(x="fecha",y=["respiradores_covid_totales","respiradores_covid_ocupados"] )
plt.title("Respiradores covid", fontdict={"family": "serif", 'weight': 'bold','size': 18})
plt.legend(bbox_to_anchor=(1.05, 1.0), loc="upper left")
plt.show()
```

**Observaciones**

* Se puede observar que existen fechas en las que los datos no son confiables, 
ya que el número de camas de los hospitales es cero. Para tener un mejor análisis, los datos correspondientes a esas fechas deberían ser eliminados. 
* Se puede ver que a partir de cierta fecha se incorporan respiradores dedicados a uso exclusivo de covid, por esta razón pasa de cero a más de cuarenta en un solo día.
